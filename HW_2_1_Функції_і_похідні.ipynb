1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp

# Визначаємо символічну змінну
x = sp.symbols('x')
```

```python
# Лінійна функція
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції 10x-5:", derivative_f_lin)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)
```

```python
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x-5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Функція лінійна, тому оцінка зростає пропорційно часу.
Похідна = 10 означає, що кожна додаткова година дає +10.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

print("Похідна функції sin(x):", derivative_f_sin)
```

```python
# робимо числові функції
f_sin_num = sp.lambdify(x, f_sin, "numpy")
df_sin_num = sp.lambdify(x, derivative_f_sin, "numpy")

# numpy-вісь
x_vals = np.linspace(0, 10, 100)
```

```python
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_sin_num(x_vals), label="f(x)=sin(x)")
plt.plot(x_vals, df_sin_num(x_vals), label="f'(x)=cos(x)")

# Додавання легенди, міток та заголовка
plt.title("sin(x) та похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна функції sin(x):", derivative_f_ln)
```

```python
# робимо числові функції
f_ln_num = sp.lambdify(x, f_ln, "numpy")
df_ln_num = sp.lambdify(x, derivative_f_ln, "numpy")

# numpy-вісь
x_vals = np.linspace(0.1, 10, 100)
```

```python
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_ln_num(x_vals), label="f(x)=ln(x)")
plt.plot(x_vals, df_ln_num(x_vals), label="f'(x)=1/x")

# Додавання легенди, міток та заголовка
plt.title("ln(x) та похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f_lin = x**2 - 5*x +1
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна функції x**2 - 5*x + 1 :", derivative_f_lin)
```

```python
# робимо числові функції
f_num = sp.lambdify(x, f_lin, "numpy")
df_num = sp.lambdify(x, derivative_f_lin, "numpy")

# numpy-вісь
x_vals = np.linspace(0, 100, 400)
```

```python
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_num(x_vals), label="f(x) = x**2 - 5*x + 1")
plt.plot(x_vals, df_num(x_vals), label="f'(x) = 2*x - 5")

# Додавання легенди, міток та заголовка
plt.title("x**2 - 5*x + 1 та похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Функція показує, що зі збільшенням обсягу виробництва загальні витрати зростають. Похідна 2x−5 показує швидкість цього зростання: чим більше виробництво, тим швидше починають рости витрати. Це означає, що при великих обсягах виробництва витрати збільшуються інтенсивніше.
