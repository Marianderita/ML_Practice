**1. Цикл for - базовий**

Використовуючи цикл `for`, виведіть числа від 1 до 7 включно використовуючи функцію range.


```python
for i in range(7):
  print(i+1)
```

**2. Цикл for по списку**

Дано список:

```python
names = ["Олег", "Анна", "Іван", "Марія"]
```


Використовуючи цикл, виведіть кожне ім'я з привітанням: "Привіт, [ім'я]!"


```python
for name in names:
  print(f"Привіт, {name}!")
```

**3. Цикл з enumerate**

Дано список оцінок:

```python
grades = [85, 92, 78, 95, 88]
```


Використовуючи `enumerate`, виведіть кожну оцінку з номером: "Студент 1: 85 балів"


```python
for i, grade in enumerate(grades):
  print(f"Студент {i+1}: {grade} балів")
```

**4. Сума парних чисел**

Напишіть програму, яка обчислює суму всіх парних чисел від 1 до 20.

```python
list_data = [x for x in range(1, 21) if x % 2 == 0]
sum_data = 0

for x in list_data:
  sum_data = sum_data + x
print(sum_data)
```

```python
sum_data = 0
for x in range(2, 21, 2):
    sum_data += x

print(sum_data)
```

```python
sum_data = sum(x for x in range(1, 21) if x % 2 == 0)
print(sum_data)
```
