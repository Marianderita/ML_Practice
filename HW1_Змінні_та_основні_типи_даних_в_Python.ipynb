**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4

print(var1 + var2) # сума
print(var1 - var2) # різниця
print(var1 * var2) # добуток
print(var1 / var2) # ділення
print(var1 // var2) # цілочисельне ділення
print(var1 % var2) # залишок від ділення
print(var1 ** var2) # піднесення до степеня
```
